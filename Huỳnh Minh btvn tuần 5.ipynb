```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from datetime import date
np.set_printoptions(precision=3, suppress=True)
pd.options.mode.use_inf_as_na = True
```

```python
#Bài 1: Lấy thông tin
housing_data=pd.read_csv('./data/vn_housing_dataset/VN_housing_dataset.csv')
housing_data.info()
housing_data.head()
```

```python
 house_price=housing_data[['Diện tích','Dài','Rộng','Giá/m2']]
#date['Ngày']=housing_data[['Ngày']]                   
#house_date=house_price.join(date)

house_price
```

```python
# Bài 2: Kiểm tra các thông tin bất thường
house_price.isna().sum()

```

```python
#loại bỏ các thông tin bất thường ở các cột
drop_nan=house_price.dropna(axis = 0, how = "any")
drop_nan
```

```python
#Kiểm tra lại xem đã xoá hết các giá trị NAn ở chưa
drop_nan.isna().any()
```

```python
#format ngày
date=housing_data[['Ngày']]   
print('Ngày trước khi đổi:\n',date)
datenew=date['Ngày'].str.replace('-','_')
print('Ngày sau khi đổi:\n',datenew)
```

```python
#format giá nhà
price=housing_data["Giá/m2"]
for value in price:
    if 'tỷ' in str(value):
        ty_to_trieu=value.replace('tỷ', 'triệu')
        print(ty_to_trieu)
pricenew=drop_nan['Giá/m2'].str.replace('tỷ', 'triệu')
print("Giá sau khi đổi: \n", pricenew)
    
```

```python
sum = pd.concat((drop_nan.drop('Giá/m2',axis=1), pricenew, datenew),axis=1)
sum.dropna(axis =0, how ='any').to_csv('bài tập tuần 5.csv', index=False)
```
